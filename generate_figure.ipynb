```python
import networkx as nx
import matplotlib.pyplot as plt
edges = [
(7, 2),
(9, 1),
(10, 1),
(6, 0),
(0, 3),
(11, 4),
(9, 2),
(6, 6),
(0, 3),
(8, 1),
(11, 2),
(4, 6),
]
G = nx.Graph()

f = plt.figure(figsize=(15, 15))


G.add_edges_from(edges)
print(nx.is_directed_acyclic_graph(G))
nx.draw(G)
f.savefig("graph.png")
plt.plot()
```
